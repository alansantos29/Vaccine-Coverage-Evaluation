### creating a varibale EstmateUS which is made of the grouping of all the  Vaccines, states andd Dimesnions in the US

```python
#creating a varibale EstmateUS which is made of the grouping of all the  Vaccines, states andd Dimesnions in the US
table1.groupby(['Vaccine','Dimension']).Estimate.mean()
EstimateUS = table1.groupby(['Vaccine', 'Dimension'])['Estimate'].agg(EstimateUS='mean').reset_index()
EstimateUS
```

### Creation of variable table1_FL, which is a table, so that only data from florida is available 

```python
# Creation of variable table1_FL, which is a table, so that only data from florida is available 
table1_FL=table1[(table1['Geography']=='Florida')]
```

### Creation of a variable EstimateFl which will later on be used to compare with the ggregated means of all the Sates in the US

```python
#Creation of a variable EstimateFl which will later on be used to compare with the ggregated means of all the Sates in the US
table1_FL.groupby(['Vaccine','Dimension']).Estimate.mean()
EstimateFL = table1_FL.groupby(['Vaccine','Dimension'])['Estimate'].agg(EstimateFL='mean').reset_index()
EstimateFL
```

```python
pd.set_option('Display.Max_row', None)
#Calculate each Row against one another, to Find the Differences - probably will have to create a table that has both results and then tley will create a 
#new column with the increase or decrease for the state of Florida
EstimateCompare= pd.merge(EstimateFL,EstimateUS, on=['Vaccine','Dimension'], how='left')
#Calculate the increase in the estimate percentage in the state of Florida compared to the rest of the US states average
EstimateCompare['Increase'] = ((EstimateCompare['EstimateFL'] - EstimateCompare['EstimateUS']) / EstimateCompare['EstimateUS']) * 100
EstimateCompare
```

```python
# Assuming you have your DataFrame 'EstimateCompare' with columns 'Vaccine', 'Dimension', and 'Increase'
plt.figure(figsize=(16, 10))  # Increase the figure size

# Create the bar chart
ax = sns.barplot(x='Vaccine', y='Increase', hue='Dimension', data=EstimateCompare)
plt.xticks(rotation=45)
plt.title('Increase by Vaccine and Dimension')
plt.xlabel('Vaccine')
plt.ylabel('Increase')

# Adjust the legend position
plt.legend(title='Dimension', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()  # Ensures that the labels and legends do not overlap
plt.show()
```

```python
# Create dataframes
df = pd.DataFrame(EstimateCompare)

# Get unique vaccine names
vaccines = df["Vaccine"].unique()

# Set the width of each bar in the grouped bar chart
bar_width = 0.25

# Iterate through each vaccine and create a grouped bar chart
for vaccine in vaccines:
    vaccine_data = df[df["Vaccine"] == vaccine]
    indices = np.arange(len(vaccine_data))
    
    # Create subplots for the grouped bar chart
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Bar chart for EstimateFL
    ax.bar(indices - bar_width, vaccine_data["EstimateFL"], bar_width, label='EstimateFL', color='b', alpha=0.7)
    
    # Bar chart for EstimateUS
    ax.bar(indices, vaccine_data["EstimateUS"], bar_width, label='EstimateUS', color='g', alpha=0.7)
    
    # Bar chart for Increase
    ax.bar(indices + bar_width, vaccine_data["Increase"], bar_width, label='Increase', color='r', alpha=0.7)
    
    ax.set_title(f'{vaccine} - All Data Columns')
    ax.set_xticks(indices)
    ax.set_xticklabels(vaccine_data["Dimension"], rotation=45, ha='right')
    ax.set_ylabel('Values')
    ax.legend()
    
    plt.tight_layout()
    plt.show()
```
